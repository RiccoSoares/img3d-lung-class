Functions to segment CT images (nii) with a range or mask

```python
import numpy as np
import nibabel as nib
import os
from glob import glob
import time
```

```python
def crop_3d_by_range(file_in, lim_inf, lim_sup, file_out):
    img_ori = nib.load(file_in)

    data_ori = img_ori.get_fdata()
    a_max = np.amax(data_ori)
    data_ori[data_ori<lim_inf] = a_max
    data_ori[data_ori>lim_sup] = a_max
    crop_img = nib.Nifti1Image(data_ori, img_ori.affine, img_ori.header, img_ori.extra, img_ori.file_map)
    nib.save(crop_img, file_out)


def crop_3d_by_mask(file_in, file_mask_in, file_out):
    img_ori = nib.load(file_in)
    img_mask = nib.load(file_mask_in)

    data_ori = img_ori.get_fdata()
    data_mask = img_mask.get_fdata()
    bool_mask = data_mask > 0
    data_ori[~bool_mask] = 0.0
    #print(data_ori.shape)
    #print(img_ori.affine)
    #print(img_ori.header)
    #print(img_ori.extra)
    #print(img_ori.file_map)
    crop_img = nib.Nifti1Image(data_ori, img_ori.affine, img_ori.header, img_ori.extra, img_ori.file_map)
    nib.save(crop_img, file_out)
```

```python
INPUT_DIR_FORMAT = "/home/guilherme/Documents/noa/cidia19/data/*/exame-pulmao/*/*/crop_by*.nii.gz"
```

```python
print('Begin: convert dicom to nifti')
# time_init = time.time()
files = glob(INPUT_DIR_FORMAT)
i = 0
for f in files:
    dir_f = os.path.dirname(f)
    name = f.split('/')[-1]
    name = dir_f + "/vidro_fosco_" + name.replace("crop_by_mask_", "")
    crop_3d_by_range(f, -700, -233, name)
    i+=1
    print("{} de {}".format(i, len(files)))
    
```
